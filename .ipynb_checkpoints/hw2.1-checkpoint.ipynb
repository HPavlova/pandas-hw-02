```python
import pandas as pd
```

```python
tmp = pd.read_html('https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C', match="(1950—2019)")
df = tmp[0].head()
df
```

```python
df.shape
```

```python
df = df.replace('—', 'NaN')
df.dtypes
```

```python
df.iloc[:,1:] = df.iloc[:,1:].astype('float64')
df.dtypes
```

```python
pd.isna(df).sum()
```

```python
df1 = tmp[0].head(-1)
df1 = df1.replace('—', 'NaN')
df1.iloc[:,1:] = df1.iloc[:,1:].astype('float64')
df1
```

```python
df1 = df1.fillna(df1.iloc[:,1:].mean())
df1
```

```python
mean_2019 = df1['2019'].mean()
df2 = df1.iloc[:,:1][df1['2019'] > mean_2019]
df2
```

```python
max_2014 = df1['2014'].max()
df3 = df1.iloc[:,:1][df1['2014'] == max_2014]
df3
```

```python
df1.plot.bar(x='Регіон', y='2019')
```
